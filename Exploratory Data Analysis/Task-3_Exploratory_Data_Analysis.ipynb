## Analysing Ship Mode

```python
df
```

```python
sales_by_shipmode = df.groupby('Ship Mode')['Sales'].sum()
ax = sales_by_shipmode.plot(kind='bar',title='Sales per Ship Mode')
ax.set_xlabel('Ship Mode')
ax.set_ylabel('Sales')
print(sales_by_shipmode)
plt.show()
```

```python
profit_by_shipmode = df.groupby('Ship Mode')['Profit'].sum()
ax = profit_by_shipmode.plot(kind='bar',title='Profits per Ship Mode')
ax.set_xlabel('Ship Mode')
ax.set_ylabel('Profit')
print(profit_by_shipmode)
plt.show() 
```

```python
quantity_by_shipmode = df.groupby('Ship Mode')['Quantity'].sum()
ax = quantity_by_shipmode.plot(kind='bar',title='Quantity sold per Ship Mode')
ax.set_xlabel('Ship Mode')
ax.set_ylabel('Quantity')
print(quantity_by_shipmode)
plt.show()
```

#### Conclusion
Among all different ship mode Standard class has the high sales,high profits and most number of sold quandities

## Analysing Segment

```python
df
```

```python
Sales_per_segment = df.groupby('Segment')['Sales'].sum()
print(Sales_per_segment)
bx=Sales_per_segment.plot(kind='bar',title='Sales per Segment')
bx.set_xlabel('Segment')
bx.set_ylabel('Sales')
plt.show()
```

```python
Profit_per_segment = df.groupby('Segment')['Profit'].sum()
print(Profit_per_segment)
bx=Profit_per_segment.plot(kind='bar',title='Profit per Segment')
bx.set_xlabel('Segment')
bx.set_ylabel('Profit')
plt.show()
```

```python
Quandity_per_segment = df.groupby('Segment')['Quantity'].sum()
print(Quandity_per_segment)
bx=Quandity_per_segment.plot(kind='bar',title='Quantity per Segment')
bx.set_xlabel('Segment')
bx.set_ylabel('Quantity')
plt.show()
```

```python
Discount_per_segment = df.groupby('Segment')['Discount'].sum()
print(Discount_per_segment)
bx=Discount_per_segment.plot(kind='bar',title='Discount per Segment')
bx.set_xlabel('Segment')
bx.set_ylabel('Discount')
plt.show()
```

#### Conclusion
In segment consumer has the high sales and generating high profit \
consumer has the high quandity of sales\
Home office sale is poor therefore profit too

## State wise Analysis

```python
Profit_per_state = df.groupby('State')['Profit'].sum()
dx=Profit_per_state.plot(kind='bar',title='Profit per States',figsize=(15,4))
dx.set_ylabel('profit')
plt.show()
```
